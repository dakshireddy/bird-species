```python
import os
import numpy as np
import random
import matplotlib.pyplot as plt

import cv2
from PIL import Image
```

```python
import os
import cv2

# Make sure directory1 is correctly defined
directory1 = "C:/AD/Test"

data = []
c = 0
labels = []

for cats in os.listdir(directory1):
    if c != 100:
        folder = os.path.join(directory1, cats)
        label = cats  # Corrected: Assigning category name to label variable
        labels.append(label)
        ch = 0
        for img_name in os.listdir(folder):  # Corrected: Renamed variable to avoid overwriting img
            if ch != 20:
                img_path = os.path.join(folder, img_name)  # Corrected: Renamed variable and used os.path.join
                img_arr = cv2.imread(img_path)
                img_arr = cv2.resize(img_arr, (224, 224))
                data.append([img_arr, label])
            else:
                break
            ch += 1
    else:
        break
    c += 1
```

```python
random.shuffle(data)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Conv2D,Conv3D,MaxPooling2D,Activation,Flatten,Dense,Dropout,Dropout,BatchNormalization
from keras.models import Sequential
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.optimizers import SGD,Adam
```

```python
import numpy as np
img_width = 224  # Replace with the actual width of your images
img_height = 224  # Replace with the actual height of your images


# Ensure that labels are within the range [0, 13]
y_train_corrected = np.clip(y_train, 0, 13)

# Define your model architecture
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(img_width, img_height, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(14, activation='softmax')  # Assuming you have 14 classes
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
model.fit(x_train, y_train_corrected, batch_size=32, epochs=25, verbose=1)
```

```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import img_to_array
import matplotlib.pyplot as plt

# Load your model
model_path = r'C:/Users/gopi naga priya/project/nnsavedmodels/chillimodelpred72.h5'
model = tf.keras.models.load_model(model_path)

# Load bird names from the file
def load_labels(file_path):
    with open(file_path, 'r') as file:
        labels = file.read().splitlines()
    return labels

# Path to the bird names file
labels_file_path = 'bird_names.txt'
labels = load_labels(labels_file_path)

def preprocess_frame(frame):
    """
    Preprocess the frame for the model prediction.
    
    Parameters:
    frame (numpy.ndarray): The captured frame from the video feed.
    
    Returns:
    numpy.ndarray: The preprocessed frame.
    """
    img = cv2.resize(frame, (224, 224))  # Resize frame to 224x224
    img = img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = img / 255.0  # Scale image values to [0, 1]
    return img

def predict_bird(frame, model, labels):
    """
    Predict the bird species from the frame.
    
    Parameters:
    frame (numpy.ndarray): The captured frame from the video feed.
    model (tf.keras.Model): The trained Keras model for prediction.
    labels (list): The list of bird species labels.
    
    Returns:
    str: The predicted bird species name.
    """
    preprocessed_frame = preprocess_frame(frame)
    pred = model.predict(preprocessed_frame)
    print(f"Model prediction (raw output): {pred}")
    
    bird_index = np.argmax(pred)
    print(f"Predicted index: {bird_index}")
    
    # Check if the predicted index is within the range of labels
    if bird_index < len(labels):
        bird_name = labels[bird_index]
    else:
        bird_name = "Unknown"
        print(f"Warning: Predicted index {bird_index} is out of range for labels.")
    
    return bird_name

# Test function to check model prediction on a sample image
def test_model_on_sample_image(image_path, model, labels):
    """
    Test the model on a sample image to verify predictions.
    
    Parameters:
    image_path (str): The path to the test image.
    model (tf.keras.Model): The trained Keras model for prediction.
    labels (list): The list of bird species labels.
    """
    img = cv2.imread(image_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    bird_name = predict_bird(img_rgb, model, labels)
    plt.imshow(img_rgb)
    plt.title(f"Predicted Bird: {bird_name}")
    plt.axis('off')
    plt.show()

# Path to a sample test image (adjust this to point to a known test image)
sample_image_path = "C:\AD\Test\Acadian_Flycatcher\Acadian_Flycatcher_0014_795607.jpg"
test_model_on_sample_image(sample_image_path, model, labels)

# Start video capture
cap = cv2.VideoCapture(0)  # 0 is usually the default camera

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    # Convert frame to RGB
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    # Predict the bird species
    bird_name = predict_bird(frame_rgb, model, labels)
    
    # Display the frame with the predicted bird name
    cv2.putText(frame, f"Predicted Bird: {bird_name}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow('Bird Species Prediction', frame)
    
    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture and close windows
cap.release()
cv2.destroyAllWindows()
```
